```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[['Attrition', 'Department']]
y_df
```

```python
# Create a list of at least 10 column names to use as X data

X_data = attrition_df[['BusinessTravel', 'DistanceFromHome', 'EnvironmentSatisfaction', 'HourlyRate', 'JobInvolvement', 'JobSatisfaction', 'NumCompaniesWorked', 'WorkLifeBalance', 'YearsSinceLastPromotion', 'PerformanceRating', 'JobRole', 'OverTime']]


# Create X_df using your selected columns
X_df = pd.DataFrame(X_data)

# Show the data types for X_df
X_df.dtypes
```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary
le = LabelEncoder()

stand_labels = X_df[['BusinessTravel', 'JobRole', 'OverTime']]
def le_encoder(labels):
    encoded_labels = []
    for i in labels.columns:
        labels = le.fit_transform(stand_labels[i])
        encoded_labels.append(labels)
    result = pd.DataFrame(encoded_labels).T
    return result
```

```python
x_encoded = le_encoder(stand_labels)
x_encoded = x_encoded.rename(columns={0: 'BusinessTravel', 1: 'JobRole', 2: 'OverTime'})
x_encoded
```

```python
X_df = X_df.drop(columns={'BusinessTravel', 'JobRole', 'OverTime'})
X_df = pd.concat([X_df, x_encoded], axis=1)
X_df
```

```python
X_df.dtypes
```

```python
oh = OneHotEncoder(sparse=False)  # Use sparse=False to return a dense array

def oh_encoder(labels):
    encoded_labels = []
    encoded_columns = []
    
    for col in labels.columns:
        # Reshape each column to 2D and encode
        label_reshaped = labels[col].values.reshape(-1, 1)
        encoded = oh.fit_transform(label_reshaped)
        encoded_labels.append(encoded)
        # Add feature names to track which columns were encoded
        encoded_columns.extend([f"{col}_{category}" for category in oh.categories_[0]])
    
    # Combine all encoded columns horizontally
    result = pd.DataFrame(np.hstack(encoded_labels), columns=encoded_columns)
    return result
```

```python
y_encoded = oh_encoder(y_df)
y_encoded = y_encoded.astype(int)
y_encoded
```

```python
X = X_df
y = y_encoded
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=7)
```

```python
y_train_department = y_train[['Department_Human Resources', 'Department_Research & Development', 'Department_Sales']].to_numpy()
y_train_department
```

```python
y_test_department = y_test[['Department_Human Resources', 'Department_Research & Development', 'Department_Sales']].to_numpy()
y_test_department
```

```python
y_train_attrition = y_train[['Attrition_Yes', 'Attrition_No']].to_numpy()
y_train_attrition
```

```python
y_test_attrition = y_test[['Attrition_Yes', 'Attrition_No']].to_numpy()
y_test_attrition
```

```python
# Create a StandardScaler
ss = StandardScaler()

# Fit the StandardScaler to the training data
xscaler = ss.fit(X_train)

# Scale the training and testing data
X_trainscaled = xscaler.transform(X_train)
X_testscaled = xscaler.transform(X_test)
```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
number_features = X_train.shape[1]

# Create the input layer
input = layers.Input(shape=(number_features,), name="input")

# Create at least two shared layers
sharedlayer1 = layers.Dense(64, activation="relu", name="sharedlayer1")(input)
sharedlayer2 = layers.Dense(128, activation="relu", name="sharedlayer2")(sharedlayer1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer

# Create the hidden layer
department_layer = layers.Dense(32, activation="relu", name='department_layer')(sharedlayer2)

# Create the output layer
department_output = layers.Dense(y_train_department.shape[1], activation='sigmoid', name='department_output')(department_layer)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
attrition_layer = layers.Dense(32, activation="relu", name='attrition_layer')(sharedlayer2)

# Create the output layer
attrition_output = layers.Dense(y_train_attrition.shape[1], activation='sigmoid', name='attrition_output')(attrition_layer)
```

```python
# Create the model
model = Model(inputs=input, outputs={"department_output": department_output, "attrition_output": attrition_output})

# Compile the model
model.compile(optimizer='adam',
              loss={
                  'department_output': 'categorical_crossentropy',
                  'attrition_output': 'binary_crossentropy'
              },
              metrics={
                  'department_output': 'accuracy',
                  'attrition_output': 'accuracy'
              })

# Summarize the model
model.summary()
```

```python
# Train the model
model_fitted = model.fit(X_trainscaled, {"department_output": y_train_department, "attrition_output": y_train_attrition}, epochs=100)
```

```python
# Evaluate the model with the testing data
model_results = model.evaluate(X_testscaled, {"department_output": y_test_department, "attrition_output": y_test_attrition})
print(model_results)
```

```python
# Print the accuracy for both department and attrition
print(f"Attrition Accuracy: {model_results[4]} Department Accuracy: {model_results[3]}")
```
